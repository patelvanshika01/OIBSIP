```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
data = pd.read_csv('Unemployment_in_India.csv')
data.describe()
```

```python

# Extracting relevant columns
date_column = ' Date'
unemployment_rate_column = ' Estimated Unemployment Rate (%)'
employed_column = ' Estimated Employed'
participation_rate_column = ' Estimated Labour Participation Rate (%)'
region_column = 'Region'

# Converting date column to datetime
data[date_column] = pd.to_datetime(data[date_column])
```

```python
# Plotting bar chart of average unemployment rate by region
average_unemployment_by_region = data.groupby(region_column)[unemployment_rate_column].mean()
average_unemployment_by_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Region')
plt.show()
```

```python

# Plotting bar chart of employed population by region
employed_by_region = data.groupby(region_column)[employed_column].sum()
employed_by_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Employed Population')
plt.title('Employed Population by Region')
plt.show()
```

```python
# Plotting pie chart of labor participation rates by region
participation_by_region = data.groupby(region_column)[participation_rate_column].sum()
participation_by_region.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Labor Participation Rate by Region')
plt.show()
```
